#Imports

```python
from time import sleep
import urllib3
import re
import pandas as pd
from bs4 import BeautifulSoup
import re
import requests
import json
from ar_corrector.corrector import Corrector
```

#Reading the links file

```python
# Opening JSON file
with open('/content/FatwaLinksFile.json', 'r') as openfile:
  
    # Reading from json file
    Fatawy= json.load(openfile)
  
print(Fatawy)
print(type(Fatawy))
```

#Scraping Data

```python
Question = []
answer = []
Sites=[]
corr = Corrector()#intializing the corrector
```

```python
for everylink in Fatawy:
  
  result = requests.get(everylink,allow_redirects=False)
  src = result.content
  soup = BeautifulSoup(src,"lxml")
  #soup = BeautifulSoup(page_source, 'lxml')
  #p = str(soup.find_all('p'))
  #p = soup.find_all('p')

  article = soup.find_all('article')
  TheSite = soup.find_all('span',{'class':'sitemap'})
  #print(TheSite)
  if not TheSite or not article:
    continue
  #print(TheSite[0].text)
  Sites.append(corr.contextual_correct(TheSite[0].text).replace('الرئيسية في الفتاوى في ',''))

  ques = article[0].text
  #print(ques)

  ans = article[1].text
  #print(ans)

  Question.append(ques)

  answer.append(ans)
```

#Storing data in a csv file

```python
df = pd.DataFrame(list(zip(Question,answer,Sites)),columns =['الاسئله','الاجابات','التصنيف'])
df.to_csv('Fatawy with class.csv',index=False,header=True,encoding="UTF-8")
```
