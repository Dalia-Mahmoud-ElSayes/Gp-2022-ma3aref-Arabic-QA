```python
from simpletransformers.question_answering import QuestionAnsweringModel
from transformers import AutoConfig , BertForQuestionAnswering , BertTokenizer
import torch
```

```python
config = AutoConfig.from_pretrained("/content/drive/MyDrive/AraElectra/outputs of Augmented AraBERT (augbert_05)/bert")
model = BertForQuestionAnswering.from_pretrained("/content/drive/MyDrive/AraElectra/outputs of Augmented AraBERT (augbert_05)/bert", config=config)
```

```python
tokenizer = BertTokenizer.from_pretrained("/content/drive/MyDrive/AraElectra/outputs of Augmented AraBERT (augbert_05)/bert")
```

```python
from transformers import pipeline
qa_pipe =pipeline("question-answering",model=model,tokenizer=tokenizer)
```
